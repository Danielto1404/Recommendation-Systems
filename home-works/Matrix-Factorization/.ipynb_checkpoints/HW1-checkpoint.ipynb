### Explicit данные

Для того, чтобы преобразовать текущий датасет в Implicit, давайте считать что позитивная оценка это оценка **>=4**

```python
implicit_ratings = ratings[ratings.rating >= 4]
```

```python
implicit_ratings.head(10)
```

Удобнее работать с sparse матричками, давайте преобразуем DataFrame в CSR матрицы

```python
users = implicit_ratings["user_id"]
movies = implicit_ratings["movie_id"]
user_item = sp.coo_matrix((np.ones_like(users), (users, movies)))
user_item_t_csr = user_item.T.tocsr()
user_item_csr = user_item.tocsr()

user_item_csr
```

В качестве примера воспользуемся ALS разложением из библиотеки implicit

Зададим размерность латентного пространства равным 64, это же определяет размер user/item эмбедингов

```python
model = implicit.als.AlternatingLeastSquares(factors=64, iterations=1, calculate_training_loss=True)
```

В качестве loss здесь всеми любимый RMSE

```python
model.fit(user_item_t_csr)
```

Построим похожие фильмы по 1 movie_id = Истории игрушек

```python
def get_similars(item_id, model):
    similars = model.similar_items(item_id)           # [(id, score)]
    ids      = [film_id for (film_id, _) in similars]
    masks    = [movie_info['movie_id'] == x for x in ids]
    return [movie_info[mask]['name'].to_string() for mask in masks]
```

Как мы видим, симилары действительно оказались симиларами.

Качество симиларов часто является хорошим способом проверить качество алгоритмов.

Давайте теперь построим рекомендации для юзеров

Как мы видим юзеру нравится **фантастика**, значит и в рекомендациях ожидаем увидеть **фантастику**

```python
def get_user_history(user_id, implicit_ratings):
    user_mask   = implicit_ratings['user_id'] == user_id
    film_ids    = implicit_ratings[user_mask]['movie_id']
    movie_masks = [movie_info['movie_id'] == film_id for film_id in film_ids]
    return [movie_info[mask]['name'].to_string() for mask in movie_masks]
```

#### Предобработка данных

```python
from tqdm import trange

### Зафиксируем seed чтобы избежать новых результатов при каждом запуске
# np.random.seed(239)

## Сделаем сжатую матрицу user-to-item
ratings_user_item = sp.coo_matrix((ratings.rating, (ratings.user_id, ratings.movie_id)))
ratings_user_item_csr = ratings_user_item.tocsr()
ratings_item_user_csr = ratings_user_item.T.tocsr()
```

### Задание 1. Не использую готовые решения, реализовать SVD разложение используя SGD на explicit данных


```python
class StochasticGradientDescentSVD:
    def __init__(self, iterations=1e8, factors=64, learning_rate=1e-2, alpha=1e-2, beta=1e-2, calculate_training_loss=True):
        self.iterations = np.int(iterations)
        self.learning_rate = learning_rate
        self.factors = factors
        self.alpha = alpha ## regularization parameter for user factorization
        self.beta = beta   ## regularization parameter for item factorization
        self.calculate_training_loss = calculate_training_loss
        self.user_factorization = None
        self.item_factorization = None

    def fit(self, user_to_item):
        n_users, n_items = user_item.shape
        factors = self.factors

        ## Uniformly initializes the matrix with numbers from 0 to 1 / sqrt(dimension)
        self.user_factorization = np.random.rand(n_users, factors) / np.sqrt(factors)
        self.item_factorization = np.random.rand(n_items, factors) / np.sqrt(factors)
        
        nonzero_elements = user_item.nonzero()
        n_nonzero = user_item.count_nonzero()
        
        def random_indices() -> (int, int):
            index = np.random.randint(low=0, high=n_nonzero)
            rows, columns = nonzero_elements
            return rows[index], columns[index]

        def update_vector(old_vector, gradient_vector, regularization_value, err) -> np.array:
            return old_vector * (1 - self.learning_rate * regularization_value) - 2 * err * gradient_vector

        tqdm_progress_range = trange(self.iterations)
        for it in tqdm_progress_range:
            i, j = random_indices()

            user_vector = self.user_factorization[i]
            item_vector = self.item_factorization[j]

            predicted = np.dot(user_vector, item_vector)
            error = predicted - user_to_item[i, j]

            self.user_factorization[i] = update_vector(old_vector=user_vector,
                                                       gradient_vector=item_vector,
                                                       regularization_value=self.alpha,
                                                       err=error)

            self.item_factorization[j] = update_vector(old_vector=item_vector,
                                                       gradient_vector=user_vector,
                                                       regularization_value=self.beta,
                                                       err=error)

            if it % 10_000 == 0 and self.calculate_training_loss:
                self.rmse(user_to_item=user_to_item,
                          tqdm_range=tqdm_progress_range)


    def rmse(self, user_to_item, tqdm_range, n_elements=1000):

        rows, columns = user_to_item.nonzero()
        n_nonzero = user_to_item.count_nonzero()

        indices = np.random.randint(low=0, high=n_nonzero, size=n_elements)

        def get_error(idx):
            i, j = rows[idx], columns[idx]
            predicted = self.user_factorization[i] @ self.item_factorization[j]
            actual = user_to_item[i, j]
            return (predicted - actual) ** 2

        loss = np.fromiter(map(get_error , indices),
                           np.float64)

        rmse_loss = np.sqrt(np.mean(loss))
        tqdm_range.set_postfix( {'RMSE loss' : rmse_loss} )



model = StochasticGradientDescentSVD(100_000)
model.fit(ratings_user_item_csr)
```
